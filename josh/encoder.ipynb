```python
import autoencoders as ae
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = ae.load_images('training')
```

```python
x_train, x_val = ae.norm_split_images(X)
```

```python
encoder, decoder, autoencoder = ae.build_autoencoder(x_train[1].shape, latent_dim = 32)
```

```python
autoencoder.compile(loss = 'mse', optimizer = 'adamax')
```

```python
autoencoder.fit(x_train, x_train, epochs = 20, validation_data = (x_val,x_val))
```
